## Q1. Import the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q2.import csv file Diwali Sales Data.csv

```python
df = pd.read_csv("Diwali_Sales_Data.csv", encoding='ISO-8859-1')
```

## Q6. Drop unrelated/blank columns 'Status' and 'unnamed1'

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

## Q8. Calculate the mean value of Age, Orders and Amount by using describe function

```python
df[['Age', 'Orders', 'Amount']].describe().loc['mean']
```

## Q9. Plot a bar chart for Gender and it's count 

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## Q10. Plot a bar chart for gender vs total amount 

```python
gender_amount = df.groupby('Gender')['Amount'].sum().reset_index()
sns.barplot(x='Gender', y='Amount', data=gender_amount)
plt.title('Gender vs Total Amount')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.show()
```

## Q11. Plot the countplot for Age Group with hue and Gender

```python
sns.countplot(x='Age Group', hue='Gender', data=df)
plt.title('Age Group Count by Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

## Q12. Draw the plot for total number of orders from top 10 states 

```python
top_states_orders = df.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(x='State', y='Orders', data=top_states_orders)
plt.title('Top 10 States by Total Number of Orders')
plt.xlabel('State')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.show()
```

## Q13.Draw the plot to plot total amount/sales from top 10 states

```python
top_states_amount = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
sns.barplot(x='State', y='Amount', data=top_states_amount)
plt.title('Top 10 States by Total Sales Amount')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

## Q14. Draw the plot to show the marital status 

```python
sns.countplot(x='Marital_Status', data=df)
plt.title('Marital Status Count')
plt.xlabel('Marital Status (0 = Unmarried, 1 = Married)')
plt.ylabel('Count')
plt.show()
```

## Q15. Draw the plot to show the top 10 occuations

```python
top_occupations = df['Occupation'].value_counts().head(10).reset_index()
top_occupations.columns = ['Occupation', 'Count']
sns.barplot(x='Occupation', y='Count', data=top_occupations)
plt.title('Top 10 Occupations')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## Q16. Draw the countplot to print the top 10 categories 

```python
top_categories = df['Product_Category'].value_counts().head(10).reset_index()
top_categories.columns = ['Product_Category', 'Count']
sns.barplot(x='Product_Category', y='Count', data=top_categories)
plt.title('Top 10 Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

## Q17. What is the prodct ID of the top 10 most sold products.

```python
top_products = df.groupby('Product_ID')['Orders'].sum().sort_values(ascending=False).head(10).reset_index()
top_products.columns = ['Product_ID', 'Total_Orders']
top_products
```
